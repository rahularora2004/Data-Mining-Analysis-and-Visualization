Data Analysis of Wine Dataset

Importing modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Reading CSV file

```python
data=pd.read_csv("wine.csv",names=['Alcohol','Malicacid','Ash','Alcalinity_of_ash','Magnesium','Total_phenols','Flavanoids','Nonflavanoid_phenols','Proanthocyanins','Color_intensity',"Hue","0D280_0D315_of_diluted_wines","Proline"])
```

data.head() : First 5 rows of DataFrame

```python
print("First 5 rows")
print(data.head())
```

data.tail() : Last 5 rows of DataFrame

```python
print("Last 5 rows")
print(data.tail())
```

Number of rows and columns

```python
num_rows,num_cols=data.shape
print("\n\n")
print("The data has {} rows and {} columns".format(num_rows, num_cols))
```

data.info() : DataFrame's information

```python
print("\n\nDataFrame's information")
print(data.info())
```

Summary statistics

```python
print("\n\nSummary statistics")
print(data.describe())
```

Checking null values

```python
print("\n\n Checking null values")
print(data.isnull().any())
print(data.isnull().sum())
```

Checking duplicated values

```python
print(data.duplicated())
```
